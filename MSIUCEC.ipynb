```python
import pandas as pd
from google.colab import files
import matplotlib.pyplot as plt

import io
```

```python
raw_file=files.upload()
```

```python
raw_data = pd.read_csv(io.BytesIO(raw_file['Uterine_Corpus_Endometrial_Carcinoma (4).csv']))
raw_data
```

cleaning the age to remove all entries that do not have an age

```python
raw_data.dropna(subset = ["Diagnosis Age"], inplace = True)
ac_data = raw_data.copy()
ac_data
```

Getting the median age of the people that were diagnosed with UCEC and compare that to the age of the median MSI

```python
median_age=ac_data["Diagnosis Age"].median()
median_age
```

average age by race

```python
average_age_by_race = ac_data.groupby("Race Category")["Diagnosis Age"].mean()
ax=average_age_by_race.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('Race')
plt.ylabel('Age')
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

what is the most common form of uterine cancer detected

```python
cancer_freq=ac_data['Cancer Type Detailed'].value_counts()
cancer_freq.plot(kind = 'bar')
plt.xlabel('Cancer Type')
plt.ylabel('Frequency')
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.show
```

what is the percentage of patients that survived

```python
survived=ac_data["Overall Survival Status"].value_counts()
print("The average Death rate is 19.8% ",survived[1]/survived[0])
```

what was the average age of those who died

```python
passed=ac_data.groupby(["Overall Survival Status"])["Diagnosis Age"].mean()
passed[1]
```
